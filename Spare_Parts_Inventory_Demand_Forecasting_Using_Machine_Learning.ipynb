## ***5. Hypothesis Testing***

### Hypothetical Statement - 1

####  Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

groups = [
    group['current_km_reading'].dropna()
    for name, group in df.groupby('vehicle_model')
]

f_stat, p_value = f_oneway(*groups)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

### Hypothetical Statement - 2

#### Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%d-%m-%y')
df['job_card_date'] = pd.to_datetime(df['job_card_date'], format='%d-%m-%y')

invoice = df['invoice_date'].apply(lambda x: x.toordinal())
jobcard = df['job_card_date'].apply(lambda x: x.toordinal())

corr, p_value = pearsonr(invoice, jobcard)

print("Correlation:", corr)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

### Hypothetical Statement - 3

####  Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import chi2_contingency

contingency = pd.crosstab(
    df['vehicle_model'],
    df['invoice_line_text']
)

chi2, p_value, dof, expected = chi2_contingency(contingency)

print("Chi-Square Statistic:", chi2)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define the MinMaxScaler object
scaler = MinMaxScaler()
```

```python
# Scale features (X) using the 'scaler' (MinMaxScaler) instance
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a separate scaler for the target variable (y)
y_scaler = MinMaxScaler()

# Handle infinity values in y_train and y_test before scaling
# These values were likely introduced by the transformation applied to 'current_km_reading'
import numpy as np

# Replace inf with NaN in y_train and then fill NaN with the median of finite values
y_train_cleaned = y_train.replace([np.inf, -np.inf], np.nan)
y_train_filled = y_train_cleaned.fillna(y_train_cleaned.median())

# Replace inf with NaN in y_test and then fill NaN with the median of finite values
y_test_cleaned = y_test.replace([np.inf, -np.inf], np.nan)
y_test_filled = y_test_cleaned.fillna(y_test_cleaned.median())

# Scale the target variable (y), reshaping to 2D as required by the scaler
y_train_scaled = y_scaler.fit_transform(y_train_filled.values.reshape(-1, 1))
y_test_scaled = y_scaler.transform(y_test_filled.values.reshape(-1, 1))
```

## ***7. ML Model Implementation***

### ML Model - 1   Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

lr = LinearRegression()

lr.fit(X_train_scaled, y_train_scaled)

y_pred = lr.predict(X_test_scaled)
```

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

```python
# Visualizing evaluation Metric Score chart
mae = mean_absolute_error(y_test_scaled, y_pred)
mse = mean_squared_error(y_test_scaled, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_scaled, y_pred)
```

```python
import matplotlib.pyplot as plt

scores = [mae, rmse, r2]
labels = ['MAE','RMSE','R2']

plt.bar(labels, scores)
plt.title("Linear Regression Performance")
plt.show()
```

#### 2. Cross- Validation & Hyperparameter Tuning

```python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# Fit the Algorithm

# Predict on the model
from sklearn.model_selection import cross_val_score

cv = cross_val_score(
    lr,
    X,
    y,
    cv=5,
    scoring='r2'
)

print(cv)
print(cv.mean())
```
